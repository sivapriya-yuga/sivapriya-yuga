```python
df_uk=df_online[df_online["Country"]=="United Kingdom"]
df_uk
```

```python
mybasket=df_uk.groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")
mybasket
```

```python
def myencode(x):
  if x<=0:
   return 0
  if x>=1:
    return 1
mybasket_sets=mybasket.applymap(myencode)
mybasket_sets.drop("POSTAGE",inplace=True,axis=1)
```

**APRIORI** **ANALYSIS**

```python
my_frequent_items=apriori(mybasket_sets,min_support=0.07,use_colnames=True)
my_frequent_items
```

The ASSOCiation rule

```python
my_rules=association_rules(my_frequent_items,metric="confidence",min_threshold=1)
print(my_rules)
```

Analysing Germany

```python
df_ger=df_online[df_online["Country"]=="Germany"]
```

```python
# analysing germany transactions
my_basket=df_ger.groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")
my_basket
```

```python
def myencode(x):
    if x<=0: # NEGATIVE
        return 0
    if x>=1:
        return 1

my_basket_sets=my_basket.applymap(myencode)

# remove postage column as it is not required
my_basket_sets.drop("POSTAGE",inplace=True,axis=1)
```

```python

myfrequent_items=apriori(my_basket_sets,min_support=0.07,use_colnames=True)
myfrequent_items
```

```python
my_rules=association_rules(myfrequent_items,metric="lift",min_threshold=1)
print(my_rules)
```
