```python
df_data=pd.read_csv("Car_sales.csv")
df_data
```

```python
# @title Sales vs Price over time

import matplotlib.pyplot as plt
plt.scatter(df_data['__year_resale_value'], df_data['Sales_in_thousands'], c=df_data['Price_in_thousands'],alpha=0.5)
plt.xlabel('Year')
plt.ylabel('Sales in Thousands')
plt.title('Sales vs Price over time')
```
