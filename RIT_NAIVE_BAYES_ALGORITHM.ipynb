```python
x=df.drop(['income'],axis=1)
x
```

```python
y=df['income']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
xtrain
```

```python
!pip install category_encoders
import category_encoders as ce
```

```python
encoder=ce.OneHotEncoder(cols=['workclass','education','marital_status','occupation','relationship','race','sex','native_country'])
```

```python
encoder
```

```python
xtrain=encoder.fit_transform(xtrain)
xtest=encoder.transform(xtest)
```

```python
xtrain
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(xtrain,ytrain)
```

```python
ypred=gnb.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(ytest,ypred))
```

```python
print(gnb.score(xtrain,ytrain))
```

```python
print(gnb.score(xtest,ytest))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```
