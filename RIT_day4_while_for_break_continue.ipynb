```python
fact=1
n=int(input("Enter a number"))
if n==0 :
  print("Factorial of 0 is 1")
if n>0:
  for i in range(1,n+1): #n=0 1,0
    fact=fact*i
    print("Factorial of ",i," is ",fact)
else:
  print("Please Enter a positive number")
```
