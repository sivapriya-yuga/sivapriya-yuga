Changing Column names

```python
col_names=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
df_car.columns=col_names
col_names
```

```python
df_car
```

Feature `Engineering`

```python
x=df_car.drop(["class"],axis=1)
x
```

```python
y=df_car[["class"]]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.33,random_state=21)
```

```python
xtrain
```

Encoding since it is ordinal in nature

```python
!pip install category_encoders
```

```python
import category_encoders as ce
```

```python
encoder=ce.OrdinalEncoder(cols=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety'])
encoder
```

```python
xtrain=encoder.fit_transform(xtrain)
xtest=encoder.transform(xtest)
```

Using Random forest Classfier

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(random_state=0)
```

```python
rfc.fit(xtrain,ytrain)
```

```python
print(type(xtest))
```

```python
xtest
```

```python
y_pred=rfc.predict(xtest)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score with 10 decision-trees : {0:0.4f}'. format(accuracy_score(ytest, y_pred)))
```

With estimators

```python
rfc_100=RandomForestClassifier(n_estimators=100,random_state=21)
rfc_100
```

```python
rfc_100.fit(xtrain,ytrain)
```

```python
ypred=rfc_100.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score with 100 decision-trees : {0:0.4f}'. format(accuracy_score(ytest, ypred)))
```
