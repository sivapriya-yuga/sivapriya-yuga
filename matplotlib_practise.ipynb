Simple line diagram using two points x and y
x= 0,6
y=6,250

```python
import matplotlib.pyplot as plt
import numpy as np
xpoint=np.array([0,6])
ypoint=np.array([6,250])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
import pandas as pd
df_car=pd.read_csv("Car_sales.csv")
df_car
```

```python
df_car=df_car.drop(columns=["Latest_Launch"])
```

```python
df_car.columns
```

```python
df_carhead=df_car.head(20)
```

```python
df_carhead
```

LINE PLOT

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
xpoint="Manufacturer"
ypoint="Sales_in_thousands"
plt.plot(xpoint,ypoint,data=df_carhead,marker="o")
plt.title("SALES OF DIFFRENT MANUFACTURER",color="cyan")
plt.xpoint=("Manufacturer")
plt.ypoint=("Sales in thousand")
plt.show()
```

SCATTER PLOT:

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
xpoint="__year_resale_value"
ypoint="Price_in_thousands"
plt.scatter(xpoint,ypoint,data=df_carhead)
plt.title("Comparision of Resale value with Price")
plt.show()
```

# Subplots\


```python
import matplotlib.pyplot as plt
import numpy as np
x="Engine_size"
y="Horsepower"
plt.subplot(1,2,1)
plt.plot(x,y,data=df_carhead)
plt.title("Enginepower vs HOrsepower")
x1="Engine_size"
y1="Fuel_capacity"
plt.subplot(1,2,2)
plt.plot(x1,y1,data=df_carhead)
plt.title("HORSE POWER VS fuel capacity")
plt.show()
```

# **BAR PLOT**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x="__year_resale_value"
y="Sales_in_thousands"
plt.bar(x,y,data=df_carhead)
plt.title("RESALE VALUE VS SALES")
plt.show()
```
