```python
import pandas as pd
df=pd.DataFrame({'A': [2,4,6,8],
                 'B':[6,8,10,12],
                 'c':[14,16,18,20]})
df
```

```python
df['Menu']=['Mutton Briyani','Fish Fry',"Paneer Tikka",'Butter Naan']
df
```

```python
df["Gender"]=['M','F','M',"F"]
df
```

```python
df["Gender"]=df["Gender"].map({"M":0,"F":1})
df
```

```python
df["City"]=["Madurai","Dindigul","Dharmapuri","Salem"]
df
```

```python
df.groupby("City").Gender.value_counts()
```
