HEAT MAP- Heatmap is defined as a graphical representation of data using colors to visualize the value of the matrix
function--seaborn.heatmap()

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
dataset=sns.load_dataset("tips")
dataset.head()
```

```python
df_hmap=dataset.corr()
df_hmap
```

```python
cmap="tab20"
center=0
sns.heatmap(df_hmap,cmap=cmap,center=center,annot=True)
plt.title("HEat map of TIPS")
plt.show()
```

```python
df_flights=sns.load_dataset("flights")
df_flights
```

```python
df_flights=df_flights.corr()
sns.heatmap(data=df_flights,annot=True)
```

grouping and aggregate data


```python
df_tips_agg=sns.load_dataset("tips")
df_tips_agg
```

```python
df_tips_agg=df_tips_agg.groupby(["size","day"]).agg(mean_total_bill=('total_bill',"mean"))
df_tips_agg = df_tips_agg.reset_index()
df_tips_agg
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df_flight_agg = sns.load_dataset("flights")
df_flight_agg
```

```python
df_flight_agg = df_flight_agg.groupby(["year","month"]).agg(mean_passenger=pd.NamedAgg(column="passengers", aggfunc="mean"))
df_flight_agg=df_flight_agg.reset_index()
df_flight_agg
```

```python
sns.boxplot(x="month",y="mean_passenger",data=df_flight_agg)
plt.title("Passenger travel details")
plt.show()
```
